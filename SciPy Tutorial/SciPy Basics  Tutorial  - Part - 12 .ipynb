## Sparse Matrices in SciPy.


- A Sparse Matrix is a Matrix that contains mostly zero elements. Typically, if more than 50% of its elements are zero, it's considered sparse.

- In a dense matrix, most elements are non-zero. Sparse matrices reduce memory usage compared to dense matrices.

- They save memory and computational resources by storing only non-zero elements, which is crucial for large-scale data.

- Sparse matrices are used in NLP (text), recommender systems, computer graphics, scientific computing, etc.

- You can add, subtract, and multiply sparse matrices just like dense matrices, but much faster and using less memory.

- You can access individual rows, columns, or elements efficiently using sparse matrix formats like CSR.

- You can convert a sparse matrix to a dense matrix using .toarray() or .todense(), and vice versa.

- Sparse identity, diagonal, or random matrices are useful in simulations and ML models.

- Sparse matrices are used in solving large systems of linear equations (Ax = b) efficiently.

  

```python
#1) Definition of Sparse Matrix.

import numpy as np

dense_matrix = np.array([
    [0, 0, 3],
    [0, 0, 0],
    [4, 0, 0]
])
print("Dense matrix:\n", dense_matrix)
```

```python
#2) Contrast with Dense Matrix.

from scipy.sparse import csr_matrix

sparse_matrix = csr_matrix(dense_matrix)
print("Sparse matrix (CSR format):\n", sparse_matrix)
```

```python
#3) Sparse Matrices save memory.

from sys import getsizeof

print("Size of dense:", getsizeof(dense_matrix))
print("Size of sparse:", getsizeof(sparse_matrix))  # Smaller
```

```python
#4) Common Real-world Examples.

from sklearn.feature_extraction.text import CountVectorizer

texts = ["data science", "machine learning", "science and AI"]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

print("Sparse text matrix:\n", X)
print("Dense form:\n", X.toarray())
```

```python
#5) Sparse Matrix Arithmetic.

A = csr_matrix([[0, 0, 2], [3, 0, 0]])
B = csr_matrix([[0, 0, 5], [1, 0, 0]])

print("A + B =\n", (A + B).toarray())
```

```python
#6) Indexing and Slicing.

print("Row 0 of A:", A.getrow(0).toarray())
print("Column 2 of A:", A.getcol(2).toarray())
```

```python
#7) Conversion Between Sparse and Dense.

dense_again = sparse_matrix.toarray()
print("Converted to dense:\n", dense_again)
```

```python
#8) Creation of Special Sparse Matrices.

from scipy.sparse import identity, diags

I = identity(4)
D = diags([1, 2, 3, 4])

print("Sparse Identity:\n", I.toarray())
print("Sparse Diagonal:\n", D.toarray())
```

## Sparse Diagonal Matrix.

- A Sparse Diagonal Matrix is a square matrix where only the diagonal (or a few diagonals) contains non-zero values. All other elements are zero. It's useful when you want to save space and compute faster.

- Only Diagonal Values are Non - Zero  : Only the values on the diagonal line (top-left to bottom-right) are non-zero; the rest are all zeros.

- Saves Memory Compared to Full Matrix  :  Sparse storage saves memory by not storing the zero values.

- Supports Multiple Diagonals :  You can combine several diagonals (e.g., main, upper, lower).

- Fast for Matrix Operations  : Diagonal Matrices multiply and invert quickly compared to full Matrices.

- Diagonal Extraction and Manipulation :  You can extract or modify the diagonal easily.

- Fast Element-wise Multiplication  :  You can apply them like filters or masks to keep or scale only diagonal elements.

```python
#1) Basic Diagonal Matrix.

D = diags([1, 2, 3])
print(D.toarray())
```

## Types of Sparse Matrix Representations.

- The different types of Sparse Matrix are  : COO(Coordinate Format),  CSR(Compressed Sparse Row) and CSC(Compressed Sparse Column)

## COO(Coordinate Format) 

- COO stands for Coordinate Format. It stores the matrix by keeping three arrays:

-  data: values of non-zero elements i.e. this List stores all the non-zero numbers in the matrix.  Example: If your matrix has the numbers 5, 8, and 3 that are not zero, then:
- data = [5, 8, 3]

- row: row Index of each Non-Zero Value i.e. this list tells us which row each value from the data list is in.  Example: If those numbers appear in row 0, row 1, and row 2:
- row = [0, 1, 2]
   
- col: column index of each Non-Zero Value i.e. this List tells us which Column each value from the data list is in.
- col = [2, 0, 1]





- It work together :  data = [5, 8, 3],  row = [0, 1, 2] and col = [2, 0, 1] i.e. This means:
- Put 5 at position (0, 2).
- Put 8 at position (1, 0).
- Put 3 at position (2, 1).
- All the other places in the matrix are assumed to be 0.

## CSR – Compressed Sparse Row.

- CSR stores the matrix by Compressing the Rows. It uses Three Arrays : 
    
- data: Non - Zero Values i.e. this Array holds only the values that are not Zero i.e. For Example, if your matrix is  :
- [ 5 0 0 ]

  
  [ 0 0 8 ]

  
  [ 0 3 0 ]


- Then, the non-zero values are : data = [5, 8, 3].


- indices – The column index of each non-zero value i.e. this tells us where each value is in its Row — specifically, the Column Number.
- In our example:
- 5 is in column 0.
- 8 is in column 2.
- 3 is in column 1.

- So, the indices = [0, 2, 1].



- indptr – Index pointer: where each row starts in data i.e. This array tells us: where each row starts and ends in the data Array. It helps us split the non-zero values back into their original rows. In the example:
- Row 0: 1 non-zero value → starts at index 0, ends before index 1.
- Row 1: 1 value → starts at 1, ends before 2.
- Row 2: 1 value → starts at 2, ends before 3.
- So : indptr = [0, 1, 2, 3] (There is always 1 more element than the number of rows).

## CSC – Compressed Sparse Column.

- CSC stands for Compressed Sparse Column, which is a special way of storing a sparse matrix — a matrix that has lots of zero values.

- Instead of saving Every Number in the Matrix(including zeros), we only save the important non-zero values and where they are — but in a column-wise manner.

- CSC stores the matrix by compressing the columns. It uses :
  
- data: Non - Zero Values i.e. This is just a list of all the non-zero values in the Matrix and we go Column by Column, from left to right, and collect the non-zero values.
- If the matrix is  :

[0 4 0]

  
[1 0 0]

  
[0 2 3]

  
- Then  :  Column  0  has  :  1, Column 1 has: 4, 2 and Column 2 has :  3 and So, the data = [1, 4, 2, 3].
  
 
- indices: Row Indices of each value i.e. Then  :  Column 0 has: 1,  Column 1 has: 4, 2 and  Column 2 has: 3.  So, data = [1, 4, 2, 3].
From above:
- 1 is from row 1.
- 4 is from row 0.
- 2 is from row 2.
- 3 is from row 2.
- So :  indices = [1, 0, 2, 2]


- indptr: index pointer — tells where each column starts and ends in data i.e. This says where each column starts and ends in the Data Array and  where each column’s values start and stop inside data
- From the earlier example : Column 0 starts at index 0, ends before index 1, Column 1 starts at index 1, ends before index 3 and Column 2 starts at index 3, ends before index 4. So, indptr = [0, 1, 3, 4]

```python
from scipy.sparse import csc_matrix

dense = [[0, 2, 0], [0, 0, 3], [4, 0, 0]]
csc = csc_matrix(dense)
print(csc.toarray())
```

## Uses Three Arrays: data, indices, indptr i.e. data: Non - Zero Values,  indices  :  Row Indices and  indptr: column pointer(start of each column in data)



```python
print("Data   :", csc.data)
print("Indices:", csc.indices)
print("IndPtr :", csc.indptr)
```

## Efficient for Column Access  :  Column slicing is fast.

```python
print("Column 1:", csc[:, 1].toarray())
```

## Not Optimized for Row Access  :  Accessing rows is Slower compared to CSR.

```python
print("Row 2:", csc[2].toarray())  # Still possible
```

## Matrix-Vector Multiplication(Column-wise)  :  CSC is efficient for Column-Based Computations.

```python
import numpy as np
vec = np.array([1, 2, 3])
result = csc.dot(vec)
print("Matrix * Vector =", result)
```
