## Basic Plotting with Matplotlib in SciPy.

- Plotting Basic Line Graphs  :  plt.plot() is the fundamental function for line graphs.

- Adding Titles and Labels  :  Graphs are more readable with titles and axis labels.

- Using Different Line Styles and Colors  :  You can change line styles (dotted, dashed) and colors using format strings or keyword arguments.

- Plotting Multiple Lines  :  Multiple datasets can be plotted on the same axes.

- Plotting Multiple Lines  :  Multiple datasets can be plotted on the same axes.

- Scatter Plots   :  Used for plotting individual data points without connecting lines.

- Subplots  :  Use plt.subplot() to create multiple plots in one figure.

- Bar Charts  :  Bar plots are used to compare quantities across categories.

- Histograms   :  Histograms show the distribution of a dataset.

- Saving Figures  : Plots can be saved as image files using plt.savefig().

- Customizing Ticks : You can customize tick marks and labels.

- Adding Grid : Adding grids makes it easier to read data values.

- Wide Variety of Plot Types  :  Matplotlib supports numerous plot types: line, bar, scatter, histogram, pie chart, boxplot, and more — enabling versatile data visualization.

- Integration with NumPy and SciPy  :  Matplotlib seamlessly integrates with NumPy arrays and SciPy outputs for direct plotting of mathematical and scientific data.

- Customization of Plot Elements : You can modify line styles, colors, markers, font sizes, axes, labels, ticks, and more for professional-quality plots.

- Subplots for Comparative Visualization : Use subplot() or subplots() to display multiple plots side by side or stacked for comparison.

- Colormaps and Contour Plots  : Colormaps are used to represent data intensity or magnitude, useful in heatmaps or 2D functions.

- Compatible with Pandas DataFrames  : Pandas integrates tightly with Matplotlib, allowing quick and easy visualization of DataFrames.


```python
#1)  Basic Example of Basic Plotting with Matplotlib in SciPy.

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Simple Sine Wave")
plt.show()
```

```python
#2) Plotting Basic Line Graphs.

x = np.array([0, 1, 2, 3])
y = np.array([0, 1, 4, 9])

plt.plot(x, y, label="y = x^2")
plt.legend()
plt.show()
```

```python
#3) Adding Titles and Labels.

plt.plot(x, y)
plt.title("Quadratic Function")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
#4)  Using Different Line Styles and Colors.

plt.plot(x, y, 'r--')  # Red dashed line
plt.show()
```

```python
#5) Plotting Multiple Lines.

y2 = x**3
plt.plot(x, y, label="x^2")
plt.plot(x, y2, label="x^3")
plt.legend()
plt.show()
```

```python
#6) Scatter Plots.

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
#7) Subplots.

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("x^2")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("x^3")

plt.tight_layout()
plt.show()
```

```python
#8) Bar Charts.

categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

```python
#9) Histograms.

data = np.random.randn(1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram")
plt.show()
```

```python
#10) Customizing Ticks.

plt.plot(x, y)
plt.xticks([0, 1, 2, 3])
plt.yticks([0, 5, 10])
plt.show()
```

```python
#11)  Adding Grid.


plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
#12)  Line and Bar Plot.

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(5)
y = np.array([2, 4, 6, 8, 10])

plt.plot(x, y, label="Line")
plt.bar(x, y, alpha=0.5, label="Bar")
plt.legend()
plt.title("Line and Bar Plot")
plt.show()
```

```python
#13) Integration with NumPy and SciPy : Signal Plot from SciPy.


from scipy import signal

t = np.linspace(0, 1, 500)
square_wave = signal.square(2 * np.pi * 5 * t)

plt.plot(t, square_wave)
plt.title("Square Wave (SciPy + Matplotlib)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
#14) Customization of Plot Elements :  Styled Plot.

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, linestyle='--', color='red', marker='o')
plt.title("Customized Line Plot")
plt.xlabel("X")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

```python
#15) Subplots for Comparative Visualization : Sine and Cosine.

x = np.linspace(0, 10, 100)

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))
plt.title("Sine")

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.title("Cosine")

plt.tight_layout()
plt.show()
```

```python
#16)  Colormaps and Contour Plots.

X, Y = np.meshgrid(x, x)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar()
plt.title("Contour Plot with Colormap")
plt.show()
```

```python
#17) Compatible with Pandas DataFrames.

import pandas as pd

df = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.randn(100)
})

df.plot(kind='scatter', x='A', y='B', title="Pandas + Matplotlib")
plt.show()
```

## Examples of Line Plots in SciPy.

## Basic Sine Wave.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 500)
y = np.sin(x)

plt.plot(x, y)
plt.title("Basic Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

## Multiple Lines on the Same Axes.

```python
x = np.linspace(0, 10, 400)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
plt.title("Two Trig Functions")
plt.show()
```

## Custom Line Style, Color & Markers.



```python
plt.plot(x, np.sin(x),
         color='purple', linestyle='--', linewidth=2,
         marker='o', markersize=4, markerfacecolor='white')
plt.title("Styled Line with Markers")
plt.show()
```

## Logarithmic Axes.

```python
x = np.logspace(0, 3, 400)   # 1 … 1000
plt.plot(x, np.log10(x))
plt.xscale('log')
plt.title("Log-Scale X-Axis")
plt.xlabel("x (log scale)")
plt.ylabel("log₁₀(x)")
plt.show()
```

## Heart Rate Monitor.

```python
time = np.linspace(0, 10, 100)
heart_rate = 70 + 5 * np.sin(time)
error = 2

plt.plot(time, heart_rate, label="Heart Rate", color='blue')
plt.fill_between(time, heart_rate - error, heart_rate + error, alpha=0.2)
plt.title("Heart Rate Over Time")
plt.xlabel("Time (minutes)")
plt.ylabel("BPM")
plt.legend()
plt.show()
```

## Scatter Plots.


- A scatter plot is a graph of dots, where each dot represents one observation or data point. Unlike line plots, the dots are not connected.

- Each dot represents a pair of values(like height and weight of a person).

- You plot many dots, and then look for patterns(like a trend going upward or downward).

- It is used for Showing relationships or correlations between two variables.

- It is used forSpotting clusters, outliers, or patterns in data.

-  It is used in data science, machine learning, and statistics.

-  Uses Dots to Represent Data  :  Each dot is one observation with an x and y value.

- Shows Relationships Between Two Things  :  Like height vs weight, or price vs sales.

- Great for Spotting Patterns  :  You can easily see trends like “as X increases, Y also increases.”

- Can Show Clusters  :  If data groups together, you’ll see it visually.

- Good for Detecting Outliers  : Outliers appear as dots far away from the others.

- Customizable Markers  :  Change the size, shape, and color of dots.
 
- Can Use Colors to Show Third Variable  :  Use color(and even size) to represent extra information.

- Easy to Plot Large Datasets   :  Thousands of dots can still be visualized clearly.

- Supports Tooltips in Advanced Plots  :  In interactive environments, hover info can be shown.

- Used in Machine Learning  :  To understand and visualize feature relationships.

- It’s used to show the relationship between two variables.

- Each dot is placed using an x-value and a y-value from your data.

- The points are not connected by lines — just scattered on the graph.

- Scatter plots help you spot patterns, trends, and correlations.

- For example, a rising pattern shows a positive relationship (as x goes up, so does y).

- A downward pattern shows a negative relationship (as x goes up, y goes down).

- If the dots are random with no pattern, it shows no correlation.

- You can change the color, size, and shape of the dots to show more information.

- Common in science, statistics, and machine learning.

- Ideal for finding outliers — unusual points that stand out from the rest.

- Great for exploratory data analysis to understand how variables interact.

##  Examples of Pie Charts in SciPy.

## Market Share.

```python
import matplotlib.pyplot as plt
labels = ['Brand A', 'Brand B', 'Brand C', 'Brand D']
sizes = [40, 30, 20, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title("Market Share")
plt.show()
```

## Department Spending.

```python
import matplotlib.pyplot as plt
labels = ['HR','IT','Marketing','Sales']
spend = [20, 35, 25, 20]
explode = (0, 0.1, 0, 0)
plt.pie(spend, labels=labels, explode=explode, autopct='%1.0f%%', shadow=True)
plt.title("Dept Budget")
plt.show()
```

##  Browser Usage.

```python
import matplotlib.pyplot as plt
sizes = [67, 15, 10, 8]
labels = ['Chrome','Safari','Firefox','Edge']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', pctdistance=0.8, textprops={'fontsize': 12})
plt.title("Browser Usage")
plt.show()
```

## Election Results.

```python
import matplotlib.pyplot as plt
labels = ['Party A','Party B','Party C']
votes = [330, 220, 50]
plt.pie(votes, labels=labels, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.title("Election Vote Share")
plt.show()
```

##  Sales Breakdown.

```python
import matplotlib.pyplot as plt
categories = ['Electronics','Clothing','Groceries']
shares = [45, 30, 25]
plt.pie(shares, labels=categories, autopct='%1.1f%%', labeldistance=1.05)
plt.title("Sales Composition")
plt.show()
```

## Time Distribution.

```python
import matplotlib.pyplot as plt
tasks = ['Work','Commute','Sleep','Leisure']
hours = [9, 2, 8, 5]
plt.pie(hours, labels=tasks, autopct='%1.0f%%', colors=['#8fd9b6','#d49f6a','#7597de','#d65c5c'])
plt.title("Daily Time Use")
plt.show()
```

## Donation Breakdown.

```python
import matplotlib.pyplot as plt

# Data
causes = ['Education', 'Health', 'Environment', 'Poverty']
donations = [50, 20, 15, 15]


named_colors = ['orangered', 'darkorange', 'gold', 'yellowgreen']

plt.pie(
    donations,
    labels=causes,
    autopct='%d%%',
    colors=named_colors,     # List of named colors
    startangle=180
)

plt.title("Donations by Cause")
plt.show()
```
