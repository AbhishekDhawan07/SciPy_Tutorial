## Broadcasting in SciPy.

- Broadcasting describes how NumPy and SciPy handle arithmetic operations between arrays of different shapes by virtually expanding the smaller arrays without actually copying data.


- Two shapes are compatible when, from the last dimension backwards, they are equal or one of them is 1.

- Broadcasting compares array shapes from the rightmost side. Missing dimensions are assumed to be 1.

- Scalars can broadcast to any shape since they are considered to have shape ().

- Broadcasting can handle operations involving more than 2 dimensions using the same rules.

- All NumPy/SciPy ufuncs (like np.sin, np.exp, etc.) support Broadcasting.

- Broadcasting happens automatically during operations; no manual reshaping is required.

- Boolean masks use broadcasting to filter or set values.

- Many SciPy modules (like scipy.special, scipy.optimize, scipy.stats) accept arrays with broadcasting.

- If shapes are incompatible, NumPy raises a ValueError.

- You can predict the shape of the output using the broadcasting rules without execution.

- No need to replicate data manually. Broadcasting simulates large arrays efficiently.

- Operations are concise and more readable, without needing nested loops.

- Broadcasted operations are implemented in low-level C and avoid Python loops.

- Operations like normalization, activation functions, matrix manipulation use broadcasting heavily.

- Mathematical operations like outer product, gradients, loss functions can be written naturally.

- Avoids manual reshaping, reducing the chance of index errors or memory issues.

- You can easily shift or scale arrays using scalars or constants.

```python
#1) Basic example of broadcasting.

import numpy as np
a = np.array([1, 2, 3])
b = 2
print(a + b)  # Output: [3 4 5]
```

```python
#2) Basis of Shape Compatibility.

a = np.array([[1, 2], [3, 4]])  # shape (2, 2)
b = np.array([[10], [20]])     # shape (2, 1)
print(a + b)  # shape (2, 2) => [[11, 12], [23, 24]]
```

```python
#3) Alignment from Trailing Dimensions.

a = np.ones((3, 4))
b = np.array([1, 2, 3, 4])  # shape (4,)
print(a + b)  # b is broadcast to shape (3, 4)
```

```python
#4) Broadcasting Scalar Values.

a = np.array([[1, 2], [3, 4]])
print(a + 10)  # Every element gets 10 added
```

```python
#5) Virtual Expansion(No Data Copying).

a = np.arange(3).reshape((3, 1))
b = np.array([10, 20, 30])
print(a + b)  # Result shape (3, 3)
```

```python
#6) Multi-dimensional Broadcasting.

a = np.ones((2, 1, 3))
b = np.ones((1, 4, 1))
print((a + b).shape)  # Result shape: (2, 4, 3)
```

```python
#7) Use with Universal Functions(ufuncs).

x = np.array([0, np.pi/2, np.pi])
print(np.sin(x))  # Broadcasting over 1D array
```

```python
#8) Automatic Behavior.

x = np.arange(3)
y = np.array([[10], [20], [30]])
print(x + y)  # Broadcasting shapes (3,) and (3,1) => (3, 3)
```

```python
#9)  Broadcasting with Boolean Arrays.

a = np.array([1, 2, 3, 4])
mask = a > 2
a[mask] = 0
print(a)  # Output: [1 2 0 0]
```

```python
#10) Compatibility with SciPy Functions.

from scipy.special import expit  # sigmoid function
x = np.array([[0, 1], [-1, -2]])
print(expit(x))  # shape preserved
```

```python
#11) Shape Mismatch Error.

a = np.ones((3, 4))
b = np.ones((2, 4))
# print(a + b)  # Will raise ValueError: shapes (3,4) and (2,4) not Compatible
```

```python
#12) Saves Memory.

a = np.ones((1000, 1))
b = np.ones((1, 1000))
c = a + b  # Efficiently creates (1000, 1000) without data replication
```

```python
#13) Simplifies Code.

a = np.array([1, 2, 3])
b = 10
print(a * b)  # Simpler than looping
```

```python
#14) Performance Boost.

# Vectorized with broadcasting (fast)
x = np.arange(1e6)
y = x + 10
```

```python
#15) Useful in Machine Learning.

X = np.array([[1, 2], [3, 4]])
mean = X.mean(axis=0)
X_norm = X - mean  # mean is broadcast
```

```python
#16) Cleaner Mathematical Expression.

x = np.array([1, 2, 3])
y = np.array([[1], [2], [3]])
print(x * y)  # Outer product
```

```python
#17) Reduces Bugs.

weights = np.array([0.2, 0.5, 0.3])
data = np.array([[10], [20], [30]])
# Properly aligned with broadcasting
result = data * weights
```

```python
#18) Works Well with Broadcasting-Aware Libraries.

from scipy.special import softmax
x = np.array([[1.0, 2.0], [2.0, 3.0]])
print(softmax(x, axis=1))  # Broadcasting-friendly
```

```python
#19) Broadcasting with Constants.

a = np.array([10, 20, 30])
print((a - a.min()) / (a.max() - a.min()))  # Normalize
```

## Shape Alignment from Trailing Dimensions. 


- NumPy aligns the shapes of arrays starting from the last dimension and works backward. This is crucial in broadcasting because the actual match starts from the rear.

```python
import numpy as np

a = np.ones((3, 4))
b = np.arange(4)  # shape (4,) -> treated as (1, 4)
print(a + b)      # Output shape: (3, 4)
```

## Automatic Dimension Expansion.

- NumPy implicitly expands the dimensions of smaller arrays to match the shape of the larger array using singleton dimensions(1s).


```python
a = np.array([[1], [2], [3]])  # shape (3, 1)
b = np.array([10, 20, 30])     # shape (3,) - cannot directly broadcast
b = b.reshape((1, 3))          # reshape to (1, 3)
print(a * b)                   # shape (3, 3)
```

## Broadcasting with Scalars.

- Scalars are treated as arrays of shape (), and they can broadcast with any array.

```python
a = np.array([[1, 2], [3, 4]])
print(a + 10)  # Scalar 10 is broadcast to shape (2, 2)
```

## Use in Element-wise Operations.

- Broadcasting is most commonly used in element-wise operations like addition, subtraction, multiplication, etc.

```python
a = np.array([1, 2, 3])
b = np.array([[10], [20], [30]])  # shape (3, 1)
print(a + b)  # Output shape: (3, 3)
```

## Broadcasting with Multi-Dimensional Arrays.

- Multi-dimensional broadcasting happens when arrays differ in one or more dimensions, and dimension sizes are either 1 or equal.

```python
a = np.ones((2, 1, 3))
b = np.ones((1, 4, 1))
print((a + b).shape)  # Output: (2, 4, 3)
```

## Broadcasting in Universal Functions(ufuncs).

- All NumPy and SciPy ufuncs support broadcasting. They apply operations over entire arrays while broadcasting inputs to a common shape.

```python
x = np.array([[0, 1], [2, 3]])
from scipy.special import expit  # sigmoid function
print(expit(x))  # Broadcasting-friendly
```

## Boolean Masking and Filtering with Broadcasting.

Broadcasting enables powerful boolean operations over arrays for filtering or modifying elements.

```python
a = np.array([5, 10, 15, 20])
mask = a > 10
a[mask] = 0
print(a)  # Output: [5 10 0 0]
```

## Broadcasting in Linear Algebra.

In matrix operations like normalization, dot products, or broadcasting row-wise/column-wise statistics, broadcasting simplifies operations.

```python
A = np.array([[1, 2], [3, 4]])
row_mean = A.mean(axis=1, keepdims=True)  # shape (2, 1)
print(A - row_mean)  # Broadcasting over columns
```

## Broadcasting in Vectorization.

- Broadcasting enables vectorized operations, replacing slow for-loops with efficient low-level C operations.

```python
x = np.arange(1e6)
y = x + 5  # Broadcasting adds scalar 5 efficiently
```

## Shape Inference During Broadcasting.

- You can predict the result's shape by checking if dimensions are either equal or 1. This is helpful in debugging and optimization.

```python
a = np.ones((3, 1, 5))
b = np.ones((1, 4, 1))
# Final shape will be: (3, 4, 5)
print((a + b).shape)
```

## Broadcasting in SciPy Modules.

- Modules like scipy.ndimage, scipy.stats, scipy.interpolate use broadcasting internally.

```python
from scipy.ndimage import gaussian_filter

data = np.random.rand(100, 100)
smoothed = gaussian_filter(data, sigma=[1, 2])  # Different sigma per axis (broadcast)
```

## Broadcasting in Batch Operations.

- In ML or DL, broadcasting enables batch operations across multiple samples, features, or channels.

```python
batch = np.random.rand(8, 3, 32, 32)  # 8 images, 3 channels, 32x32 pixels
mean = np.array([0.5, 0.5, 0.5]).reshape((1, 3, 1, 1))
normalized = batch - mean  # Subtract mean from each channel
```

## Features of Broadcasting in SciPy.

## Automatic Shape Adjustment.

- Broadcasting automatically adjusts the shape of arrays to be compatible for arithmetic operations without needing explicit reshaping.

```python
import numpy as np

a = np.array([1, 2, 3])         # shape (3,)
b = np.array([[10], [20]])      # shape (2,1)
print(a + b)                    # shape (2, 3) via broadcasting
```

## Memory Efficiency(No Data Copying).

- Broadcasting avoids creating large intermediate arrays. It reuses data, saving memory and improving speed.

```python
a = np.ones((1000, 1))
b = np.arange(1000)
# b is virtually expanded to (1000, 1000) without copying data
c = a * b
print(c)
```

## Scalar Compatibility.

- Scalars are Compatible with Arrays of any Shape, Automatically Broadcasting to the Array’s Shape.

```python
a = np.array([[1, 2], [3, 4]])
print(a + 10)  # Adds 10 to every element
```

## Trailing Dimension Matching.

- Arrays are aligned starting from the last dimension when comparing shapes for compatibility.

```python
a = np.ones((2, 3))
b = np.array([10, 20, 30])  # shape (3,)
print(a + b)                # b is reshaped to (1, 3)
```

## Multi-Dimensional Broadcasting.

- Broadcasting supports operations between arrays of more than two dimensions.

```python
a = np.ones((2, 1, 3))
b = np.ones((1, 4, 1))
print((a + b).shape)  # Resulting shape: (2, 4, 3)
```

## Integration with Universal Functions(ufuncs).

- All NumPy/SciPy universal functions use broadcasting to apply element-wise operations over arrays of different shapes.

```python
from scipy.special import expit  # Sigmoid function

x = np.array([[0, 1], [2, 3]])
print(expit(x))  # Applies sigmoid element-wise
```

## Simplifies Vectorized Code.

- Broadcasting makes vectorized operations intuitive and avoids writing explicit loops.

```python
x = np.array([1, 2, 3])
y = x * 10  # Vectorized multiplication
```

## Used in Conditional Logic(Boolean Arrays).

- Boolean arrays can be broadcasted for conditional indexing, filtering, or modification.

```python
a = np.array([5, 15, 25])
a[a > 10] = 0  # Broadcasting mask applied
print(a)  # Output: [5 0 0]
```

## Arithmetic, Logical, and Comparison Support.

- Broadcasting works with all element-wise operators: arithmetic (+, *), logical (&, |), and comparison (<, ==).

```python
a = np.array([1, 2, 3])
b = np.array([[3], [2], [1]])
print(a > b)  # Returns (3, 3) boolean array
```

## Dimension Preservation.

- Broadcasted Operations preserve the shape of the larger array and only expand singleton dimensions as needed.

```python
a = np.array([[1], [2], [3]])  # shape (3, 1)
b = np.array([10, 20])         # shape (2,)
print(a + b)                   # shape (3, 2)
```
