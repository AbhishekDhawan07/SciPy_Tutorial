## Stacking and Splitting in Arrays.


- Stacking means Joining Two or More Arrays along a New or Existing Axis. Common Stacking Functions: np.stack, np.hstack, np.vstack, np.dstack.

- Stacking joins two or more arrays into a single larger array.

- You can stack along rows, columns, or depth using vstack, hstack, dstack.

- np.stack() creates a new axis to combine arrays, unlike hstack or vstack..

- When stacking with np.stack(), arrays must have identical shapes.

- Stack multiple arrays(e.g., images, samples) to build a batch of data.

- Use column_stack() to stack 1D arrays into a 2D matrix (columns)..

- vstack() adds arrays as new rows in 2D.

- hstack() joins arrays side-by-side, increasing columns.

- dstack() joins 2D arrays into a 3D volume (stacking in depth.

- np.row_stack() is functionally identical to np.vstack().

- Carefully choosing the axis affects how the output is structured.


```python
import numpy as np

a = np.array([1, 2])
b = np.array([3, 4])

stacked = np.stack((a, b))
print(stacked)  # Adds a new axis
```

```python
# Combines Arrays Together.

import numpy as np
a = np.array([1, 2])
b = np.array([3, 4])
print(np.stack((a, b)))  # Stacked vertically
```

```python
# Supports Different Stack Directions.

print(np.vstack((a, b)))   # Vertical stack
print(np.hstack((a, b)))   # Horizontal stack
```

```python
# Creates New Dimensions(Axes).

c = np.stack((a, b), axis=1)
print(c)
```

```python
# All Arrays Must Be Same Shape.

a = np.ones((2,))
b = np.zeros((2,))
print(np.stack((a, b)))  
```

```python
# Useful in Image or Data Batching.

img1 = np.array([[1, 2], [3, 4]])
img2 = np.array([[5, 6], [7, 8]])
batch = np.stack((img1, img2))
print(batch.shape)  # (2, 2, 2)
```

```python
# Stacking 1D Arrays as Columns.

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.column_stack((a, b)))
```

```python
# Vertical Stack Adds Rows.

x = np.array([10, 20])
y = np.array([30, 40])
print(np.vstack((x, y)))
```

```python
# Horizontal Stack Adds Columns.

x = np.array([[1], [2]])
y = np.array([[3], [4]])
print(np.hstack((x, y)))
```

```python
# Depth Stack Adds a Third Dimension.

x = np.array([[1, 2]])
y = np.array([[3, 4]])
print(np.dstack((x, y)))
```

## Vertical Stacking i.e. np.vstack.


- np.vstack() is a NumPy function that stacks arrays vertically — i.e., row-wise — by joining them along axis 0 (the first axis). It increases the number of rows in the resulting array while keeping the columns consistent.

- It stacks input arrays vertically by adding more rows.

- If you pass 1D arrays, vstack() converts them into rows.

- It’s often used to combine data points (e.g., training samples) vertically.

- Column counts must match, otherwise it throws an error.

- Not just 1D arrays, works with already 2D matrices.

- Pass a tuple or list of multiple arrays.

- Each array becomes a row in the final array.

- Especially when output shape must be(n, features).

- If you mix 1D and 2D arrays, it converts 1D to row.


 


```python
# Stacks Arrays on Top of Each Other(Row-wise).

import numpy as np

a = np.array([1, 2])
b = np.array([3, 4])
print(np.vstack((a, b)))
```

```python
# Converts 1D Arrays into 2D Automatically.

x = np.array([10, 20])
y = np.array([30, 40])
print(np.vstack((x, y)))
```

```python
# Useful for Data Batching in ML/DL.

data1 = np.array([1, 2])
data2 = np.array([3, 4])
batch = np.vstack((data1, data2))
print(batch)
```

```python
# All Arrays Must Have Same Number of Columns.

# x = np.array([1, 2])
# y = np.array([3, 4, 5])
# np.vstack((x, y))  # ValueError
```

```python
# Works with 2D Arrays Too.

a = np.array([[1, 2]])
b = np.array([[3, 4]])
print(np.vstack((a, b)))
```

```python
# Can Stack More than Two Arrays at Once.

a = np.array([1, 2])
b = np.array([3, 4])
c = np.array([5, 6])
print(np.vstack((a, b, c)))
```

```python
# Creates Higher Number of Rows in Output.

a = np.array([9, 8])
b = np.array([7, 6])
result = np.vstack((a, b))
print(result.shape)
```

```python
# Useful for Combining Feature Vectors or Predictions.

pred1 = np.array([0.1, 0.9])
pred2 = np.array([0.8, 0.2])
combined = np.vstack((pred1, pred2))
print(combined)
```

## Depth Stacking i.e. np.dstack.


- np.dstack() stacks arrays along a third axis (depth-wise). For 1D arrays, it returns a 3D array of shape (1, n, 2); for 2D, it returns (rows, cols, arrays).

- Stacks along depth(axis=2).

- Result is always 3D.

- Each input must have same shape.

- Mixes arrays into layered form

- Preserves row and column counts.

- Result shape increases in depth.

- Access slices by depth.

```python
# Stacks along depth(axis=2).

a = np.array([[1, 2]])
b = np.array([[3, 4]])
print(np.dstack((a, b)))
```

```python
# Result is always 3D.

a = np.array([1, 2])
b = np.array([3, 4])
print(np.dstack((a, b)))  # Shape: (1, 2, 2)
```

```python
# Each input must have same shape.

a = np.ones((2, 2))
b = np.zeros((2, 2))
print(np.dstack((a, b)))
```

```python
# Mixes arrays into layered form.

a = np.array([[10, 20]])
b = np.array([[30, 40]])
print(np.dstack((a, b)))
```

```python
# Preserves row and column counts.

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(np.dstack((a, b)))
```

```python
# Result shape increases in depth.

res = np.dstack((a, b))
print(res.shape)  # (2, 2, 2)

```

```python
# Access slices by depth.

stack = np.dstack((a, b))
print(stack[:, :, 0])  # First array
```

## General Stacking i.e. np.stack.


- np.stack() joins a sequence of arrays along a new axis. Unlike hstack, vstack, etc., stack allows you to choose any axis for stacking.

- Stacks along new axis.

- In General Stacking, New dimension is added.

- Arrays must be same shape.

- Control stacking axis.

- Works on higher dimensions.

- Used in batch stacking.

- Can stack scalar values too.

- More flexible than vstack, hstack.

```python
# Stacks along new axis.

a = np.array([1, 2])
b = np.array([3, 4])
print(np.stack((a, b)))  # Axis 0 by default
```

```python
# New dimension is added.

print(np.stack((a, b), axis=1))  # Shape becomes (2, 2)
```

```python
# Arrays must be same shape.

# np.stack(([1, 2], [3]))  # ValueError
```

```python
# Control stacking axis.

print(np.stack((a, b), axis=0))
print(np.stack((a, b), axis=1))
```

```python
# Works on higher dimensions.

x = np.ones((2, 2))
y = np.zeros((2, 2))
print(np.stack((x, y), axis=0).shape)  # (2, 2, 2)
```

```python
# Used in Batch Stacking.

images = [np.random.rand(2, 2) for _ in range(3)]
print(np.stack(images).shape)  # (3, 2, 2)
```

```python
# Returns 3D from 2D inputs.

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(np.stack((a, b), axis=2))
```

```python
# Can stack scalar values too.

a = np.array(1)
b = np.array(2)
print(np.stack((a, b)))
```

```python
# More flexible than vstack, hstack.

a = np.array([1, 2])
b = np.array([3, 4])
print(np.stack((a, b), axis=-1))  # Like column stacking
```

## Column Stacking i.e. np.column_stack.

- np.column_stack() stacks 1D arrays as columns into a 2D array. Similar to np.hstack() but it first reshapes 1D arrays into columns.

- Stacks 1D arrays as columns.

- Result is always 2D.

- Useful for matrix construction.

- Reshapes 1D to (n, 1).

- Equivalent to hstack with reshape.

- Cannot stack 2D arrays with differing rows.

- Returns shape(n, m).

- Used in DataFrame-style preparation.

```python
# Stacks 1D arrays as Columns.


a = np.array([1, 2])
b = np.array([3, 4])
print(np.column_stack((a, b)))
```

```python
# Result is always 2D.

print(np.column_stack(([5, 6], [7, 8])).shape)
```

```python
# Useful for matrix construction.

ids = np.array([101, 102])
scores = np.array([95, 88])
print(np.column_stack((ids, scores)))
```

```python
# Reshapes 1D to(n, 1).

a = np.array([9, 8])
print(np.column_stack((a,)))
```

```python
# Equivalent to hstack with reshape.

np.hstack((a[:, np.newaxis], b[:, np.newaxis]))
```

```python
# Cannot stack 2D arrays with differing rows.

# np.column_stack(([1, 2], [[3], [4], [5]]))  # Error
```

```python
# Returns shape(n, m).

x = np.array([1, 2])
y = np.array([3, 4])
print(np.column_stack((x, y)).shape)  # (2, 2)
```

```python
# Used in DataFrame-style preparation.

import pandas as pd
df = pd.DataFrame(np.column_stack(([1, 2], [3, 4])), columns=['A', 'B'])
print(df)
```
