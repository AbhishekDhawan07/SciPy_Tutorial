## Inverse of the Matrix in Scipy.

- The Inverse of a Matrix is like the Reciprocal of a Number. For a Square Matrix A, its Inverse A^-1 is a Matrix that satisfies the Condition  :
A .  A^ - 1  = I,  I is the Identity Matrix (ones on the diagonal and zeros elsewhere). Not all matrices have an inverse.



```python
#  Example :  Example of Inverse of the Matrix in Scipy.

import numpy as np
from scipy.linalg import inv

A = np.array([[1, 2], [3, 4]])
A_inv = inv(A)
print("Inverse:\n", A_inv)
```

## Features of Inverse of the Matrix in Scipy.

## Only Square Matrices Have Inverses.
    
A Matrix must be Square(same Number of Rows and Columns) to have an Inverse. If the Matrix is not Square, attempting to Compute its Inverse will Result in an Error.

```python
B = np.array([[1, 2, 3], [4, 5, 6]])
# inv(B)  # This will raise an error because B is not square
```

## SciPy Provides inv() for Matrix Inversion.


- In SciPy, the function used for finding the Inverse of a Matrix is scipy.linalg.inv(). This Function is part of the scipy.linalg Module and is used for efficient linear algebra computations.

```python
from scipy.linalg import inv

C = np.array([[2, 5], [1, 3]])
C_inv = inv(C)
print(C_inv)
```

## Singular Matrices Do Not Have an Inverse.

- A Matrix is Singular if its determinant is zero. Such matrices cannot be inverted because they are not full rank (they lose dimensional information).

```python
import numpy as np
from scipy.linalg import inv, det
from scipy.linalg import LinAlgError

# Define a singular matrix (det = 0)
A = np.array([[2, 4],
              [1, 2]])  # Second row is half of the first → linearly Dependent.

# Check the determinant
print("Determinant:", det(A))  # Output: 0.0
```

## Determinant Can Predict Invertibility.

- If the Determinant is 0, the Matrix is not Invertible.

```python
import numpy as np
from scipy.linalg import inv, det  

N = np.array([[1, 2], [3, 6]])
print(det(N))  # Output: 0.0 → singular matrix
```

## Useful for Solving Linear Systems (Ax = b).


- Although solve() is better, you can use the inverse to find x in linear systems

```python
A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])
x = inv(A) @ b
print(x)
```

```python
from scipy.linalg import solve
x2 = solve(A, b)
print(np.allclose(x, x2))  # True
```

```python
b2 = np.array([[9], [8]])
x3 = inv(A) @ b2
print(x3)
```

## 10 - 20 Examples of Codes of Inverse of the Matrix in SciPy.

## Basic 2×2 Matrix Inversion.

```python
import numpy as np
from scipy.linalg import inv

A = np.array([[1, 2], [3, 4]])
print(inv(A))
```

## Identity Matrix (Inverse is itself).

```python
I = np.eye(3)
print(inv(I))
```

## Diagonal Matrix Inverse.

```python
D = np.diag([2, 4, 8])
print(inv(D))  # Inverse is 1/2, 1/4, 1/8 on diagonal
```

## Check Product A × A⁻¹ = I.

```python
A = np.array([[4, 7], [2, 6]])
A_inv = inv(A)
print(np.allclose(A @ A_inv, np.eye(2)))  # Should be True
```

## Inverse of Matrix with Random Elements.

```python
np.random.seed(0)
R = np.random.rand(3, 3)
print(inv(R))
```

## Solve Ax = b using Inverse.

```python
A = np.array([[2, 3], [1, 2]])
b = np.array([8, 5])
x = inv(A) @ b
print(x)
```

## Inverse of the Matrix with Complex Numbers.

```python
C = np.array([[1+2j, 3], [4, 5+1j]])
print(inv(C))
```

## Using scipy.linalg.det() to Check Invertibility.

```python
from scipy.linalg import det

A = np.array([[1, 2], [2, 4]])
print(det(A))  # 0 → Singular
```

## Check if Inverse of Inverse is Original.

```python
A = np.array([[3, 1], [4, 2]])
A_inv = inv(A)
print(np.allclose(inv(A_inv), A))
```

## Inverse of the Matrix with Negative Elements.



```python
N = np.array([[1, -2], [-3, 4]])
print(inv(N))
```
