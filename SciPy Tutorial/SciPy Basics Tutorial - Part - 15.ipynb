## Matrix Multiplication in SciPy.


- Matrix Multiplication in SciPy is a Binary Operation that takes a Pair of Matrices and Produces another Matrix.

- It is only defined when the Number of Columns in the First Matrix equals the Number of Rows in the second.



```python
# Example 1  : Basic Example of Matrix Multiplication in SciPy.

import numpy as np
from scipy import linalg

# Define two matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])

# Matrix multiplication using @
C = A @ B
print("A @ B =\n", C)

# Alternatively, using scipy.linalg.blas
from scipy.linalg.blas import dgemm
D = dgemm(alpha=1.0, a=A, b=B)
print("Using scipy.linalg.blas.dgemm:\n", D)
```

## Dense Matrix Multiplication.

- Dense Matrices are regular NumPy Arrays(np.array) where most or all Elements are Non - Zero. 
    
- Dense Matrix Multiplication is the Standard Matrix Product, performed using either    :   The @ Operator(Python 3.5+),  np.dot() or scipy.linalg.blas for low-level control.

- SciPy uses optimized BLAS(Basic Linear Algebra Subprograms) libraries for High Performance.



```python
# Example 1  :  Simple 2×2 Dense Matrix Multiplication.

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(A @ B)
```

```python
# Example 2 : Using np.dot().

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(np.dot(A, B))
```

```python
# Example 3  :  Using scipy.linalg.blas.dgemm.

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(dgemm(alpha=1.0, a=A, b=B))

```

## Features of Dense Matrix Multiplication in SciPy.

## Simple and Readable Syntax.

- You can perform matrix multiplication using  : @ Operator(Pythonic), np.dot() and np.matmul().

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(A @ B)           # Using @ operator
print(np.dot(A, B))    # Using np.dot()
print(np.matmul(A, B)) # Using np.matmul()
```

## Compatible with Vectors and Higher Dimensions.

- Dense Matrices can be Multiplied with 1D and Higher - Dimensional Arrays.

```python
v = np.array([1, 2])
print(A @ v)  # Matrix-vector multiplication

A3D = np.random.rand(5, 2, 3)
B3D = np.random.rand(5, 3, 4)
C3D = np.matmul(A3D, B3D)
print(C3D.shape)  # (5, 2, 4)
```

## Supports Float, Integer, and Complex Data - Types.

- Dense Multiplication works with various Data -  Types(int, float, complex).

```python
# Float32
Af = np.array([[1, 2], [3, 4]], dtype=np.float32)
Bf = np.array([[5, 6], [7, 8]], dtype=np.float32)
print(Af @ Bf)

# Complex numbers
Ac = np.array([[1+2j, 2], [3, 4+1j]])
print(Ac @ Ac)
```

## Supports Transpose Multiplication.

- Matrix Transposition is often needed for Computing Symmetric Products like  A^T.A or A.A^T.


```python
X = np.random.rand(3, 2)
print(X.T @ X)  # 2x2 matrix
print(X @ X.T)  # 3x3 matrix
```

## Power and Exponential of Matrices.


- Use scipy.linalg.matrix_power() or scipy.linalg.expm() for Advanced Linear Algebra Operations.

```python
import numpy as np
from scipy.linalg import expm
from numpy.linalg import matrix_power  # Correct import

A = np.array([[2, 0], [0, 3]])
print(matrix_power(A, 3))  # A^3

B = np.array([[0, 1], [-2, -3]])
print(expm(B))  # Matrix Exponential
```

## Broadcast and Batch Multiplication.


- Matrix Multiplication supports Broadcasting and Batched Operations across Arrays.

```python
X = np.random.rand(64, 128)
w = np.random.rand(128)
print(X @ w)  # 64 x 1 vector

A = np.random.rand(10, 4, 5)
B = np.random.rand(10, 5, 3)
print(np.matmul(A, B).shape)  # (10, 4, 3)
```

## Chain Multiplication and Associativity.

- Multiple Matrices can be Multiplied in sequence using Chained @ Operators.

```python
A = np.random.rand(2, 2)
B = np.random.rand(2, 2)
C = np.random.rand(2, 2)

result = A @ B @ C
print(result)
```

## Compatibility with Linear Algebra Solvers.


- Dense Matrix Multiplication works with other SciPy Functions like LU Decomposition and System Solvers.

```python
from scipy.linalg import solve

A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])
x = solve(A, b)
print(x)  # Solves Ax = b
```

## Sparse Matrix Multiplication.

- A Sparse Matrix is a Matrix where most values are Zero. Instead of Storing all Values, SciPy stores only the Non - Zero Values to save Memory and Speed up Calculations.

```python
# Example : Basic Example of Sparse Matrix Multiplication.

from scipy.sparse import csr_matrix

dense = [[0, 0, 5],
         [0, 0, 0],
         [1, 0, 0]]

sparse = csr_matrix(dense)
print(sparse)
```

## Features of Sparse Matrix Multiplication.

## Storing Large Matrices with mostly Zeros wastes Memory. Sparse Matrices are much more Efficient when only a few values are Non - Zero.


```python
import numpy as np
from scipy.sparse import random

dense = np.zeros((1000, 1000))
sparse = random(1000, 1000, density=0.01, format='csr')

print("Dense size:", dense.nbytes)
print("Sparse size (approx):", sparse.data.nbytes)
```

## SciPy allows Multiplication using @(Matrix Multiply) or .dot() Methods — just like NumPy.

```python
from scipy.sparse import csr_matrix

A = csr_matrix([[0, 2], [3, 0]])
B = csr_matrix([[1, 0], [0, 4]])

C = A @ B
print(C.toarray())
```

## You Can Multiply Sparse × Dense.


- A Sparse Matrix can also be Multiplied with a Regular NumPy Matrix. The result is a Dense Matrix.

```python
import numpy as np

dense = np.array([[1, 2], [3, 4]])
print((A @ dense))
```

## Efficient for Large Matrix Operations.

- Sparse Matrix Multiplication is very Fast and Efficient for large-scale data, like graphs or Machine Learning Models.

```python
from scipy.sparse import random

big_sparse = random(1000, 1000, density=0.005, format='csr')
v = np.ones(1000)

# Multiply with a vector
result = big_sparse @ v
print(result.shape)
```

## Transpose Multiplication is Easy.

- You can Multiply a Sparse Matrix with its Transpose using .T.

```python
A = csr_matrix([[0, 1], [2, 0]])
result = A.T @ A
print(result.toarray())
```

## Matrix × Vector Multiplication.

- You can Multiply a Sparse Matrix with a NumPy Vector for Operations like Ax = b.

```python
x = np.array([1, 2])
print(A @ x)
```
