## Data - Types in SciPy.


- In SciPy (via NumPy), data types (known as dtype) specify what kind of values an array holds—such as Integers, Floats, Complex Numbers, or Booleans.

- SciPy is built on NumPy, and all data structures used in SciPy are based on numpy.ndarray, which requires a fixed data type for all elements.

- Smaller data types consume less memory. Choosing the right type (like int8 instead of int64) can save memory in large datasets.

- SciPy supports many types via NumPy: Integers(int8, int16, int32, int64), Floating-point (float32, float64), Complex Numbers(complex64, complex128),
Boolean(bool) and Strings, Unicode, and structured types.

- You can define the data type of an array when creating it using the dtype parameter.

- You Can Convert (Cast) Between Data Types using .astype() to convert one type to another.

- Operations behave differently depending on the data type. For example, integer division truncates the result, while float division does not.

- SciPy supports complex numbers (used in signal processing) and boolean arrays (used in masking/filtering).

- You can define structured types (like records) using np.dtype for more complex data representation.

- Use .itemsize to know how many bytes a data type uses.

- Functions in scipy.optimize, scipy.fft, scipy.stats, etc., expect specific data types like float64.

- Choosing an appropriate data type (e.g., int64 instead of int8) helps avoid overflow or unexpected results.

  

```python
import numpy as np

arr = np.array([1, 2, 3])
print("Data Type:", arr.dtype)  # int64 or int32 based on system
```

```python
from scipy import linalg

matrix = np.array([[1, 2], [3, 4]], dtype=np.float64)
inv = linalg.inv(matrix)
print("Inverse:\n", inv)
```

```python
a = np.array([1, 2, 3], dtype=np.int8)
b = np.array([1, 2, 3], dtype=np.int64)
print("Memory Used (a):", a.nbytes)
print("Memory Used (b):", b.nbytes)
```

```python
int_arr = np.array([1, 2, 3], dtype=np.int16)
float_arr = np.array([1.1, 2.2], dtype=np.float64)
complex_arr = np.array([2 + 3j], dtype=np.complex128)
print(int_arr.dtype, float_arr.dtype, complex_arr.dtype)
```

```python
arr = np.array([1, 2, 3], dtype='float32')
print("Explicit Data Type:", arr.dtype)
```

```python
a = np.array([1.5, 2.5, 3.5])
b = a.astype(int)
print("Converted Array:", b)
print("New Data Type:", b.dtype)
```

```python
int_div = np.array([5]) // 2
float_div = np.array([5.0]) / 2
print("Integer Division:", int_div)
print("Float Division:", float_div)
```

```python
bool_arr = np.array([1, 0, 1], dtype=bool)
complex_arr = np.array([1 + 2j, 3 + 4j], dtype=np.complex64)
print("Boolean Array:", bool_arr)
print("Complex Array:", complex_arr)
```

```python
dt = np.dtype([('name', 'U10'), ('age', 'i4')])
people = np.array([('Alice', 30), ('Bob', 25)], dtype=dt)
print("Structured Data:", people)
print("Names:", people['name'])
```

```python
a = np.array([1], dtype=np.int32)
b = np.array([1], dtype=np.float64)
print("Size of int32:", a.itemsize)
print("Size of float64:", b.itemsize)
```

```python
from scipy.fft import fft

signal = np.array([1, 2, 3, 4], dtype=np.float64)
fft_result = fft(signal)
print("FFT Result:", fft_result)
```

```python
overflow = np.array([127], dtype=np.int8)
print("Before adding 1:", overflow)
print("After adding 1:", overflow + 1)  # Wraps around to -128
```

## Features of Data Types in SciPy.


- Fixed Data Type for Entire Array : Every element in a NumPy/SciPy array must be of the same data type, making operations fast and memory-efficient.

- SciPy supports for a Wide Range of Data Types  :   Integer types (int8, int16, int32, int64),  Float types (float32, float64),  Boolean,  Complex Numbers,  Strings and structured data.

- Explicit Control Over Data Type using dtype   : You can define the data type while creating an Array.

- Type Conversion (Casting) with .astype()  :  You can safely convert arrays to another data type using .astype().

- Memory Size Information Using .itemsize   :  Each Data - Type has a fixed memory size that can be retrieved with .itemsize.

- Better Performance via Appropriate Data Type Selection   :  Using Smaller Data - Types reduces memory usage and improves computation time.

- Data Type Enforces Mathematical Behavior    :  The type determines how Math Operations behave(e.g., float vs. integer division).

- Supports Complex Number Computation   :  SciPy supports complex data types used in linear algebra, FFT, etc.

- Boolean Data Types Useful in Filtering & Masking   :  Boolean Arrays are ideal for conditional selections in data analysis.

- Structured Data Types for Complex Records    :   You can define structured dtypes like records(used for tables or labeled data).

- Compatibility with SciPy Functions   :   Most SciPy modules (e.g., scipy.linalg, scipy.optimize) expect data in specific types like float64.

- Handles Overflow and Underflow Based on Type   :  Integer types have limits. Using a small type like int8 can cause overflow.




```python
#1) Fixed Data - Type for Entire Array.

import numpy as np

arr = np.array([1, 2, 3])
print("Data Type:", arr.dtype)
```

```python
#2) SciPy supports for a Wide Range of Data - Types.


arr1 = np.array([1, 2, 3], dtype=np.int32)
arr2 = np.array([1.5, 2.5], dtype=np.float64)
arr3 = np.array([True, False], dtype=bool)
arr4 = np.array([2 + 3j], dtype=np.complex128)

print(arr1.dtype, arr2.dtype, arr3.dtype, arr4.dtype)
```

```python
#3)  Explicit Control Over Data Type Using dtype.

arr = np.array([10, 20, 30], dtype='int16')
print("Explicit dtype:", arr.dtype)
```

```python
#4) Type Conversion (Casting) with .astype().

arr = np.array([1.2, 2.5, 3.8])
converted = arr.astype(int)
print("Converted Array:", converted)
```

```python
#5) Memory Size Information Using .itemsize.


a = np.array([100], dtype=np.int32)
print("Bytes per element:", a.itemsize)
```

```python
#6) Better Performance via Appropriate Data Type Selection.


small = np.array([1, 2, 3], dtype=np.int8)
large = np.array([1, 2, 3], dtype=np.int64)

print("Small dtype size:", small.nbytes)
print("Large dtype size:", large.nbytes)
```

```python
#7) Data  - Type Enforces Mathematical Behavior.

int_result = np.array([5], dtype=int) // 2
float_result = np.array([5.0], dtype=float) / 2

print("Integer division:", int_result)
print("Float division:", float_result)
```

```python
#8) Supports Complex Number Computation.

complex_arr = np.array([3 + 4j, 5 + 6j], dtype=np.complex128)
print("Complex Array:", complex_arr)
print("Magnitude:", np.abs(complex_arr))
```

```python
#9) Boolean Data -  Types Useful in Filtering & Masking.

arr = np.array([10, 15, 20])
mask = arr > 10
print("Mask:", mask)
print("Filtered:", arr[mask])
```

```python
#10) Structured Data Types for Complex Records.

dt = np.dtype([('name', 'U10'), ('score', 'f4')])
students = np.array([('Alice', 88.5), ('Bob', 92.3)], dtype=dt)

print("Names:", students['name'])
print("Scores:", students['score'])
```

```python
#11) Compatibility with SciPy Functions.

from scipy.optimize import minimize

fun = lambda x: x**2 + 2*x + 1
res = minimize(fun, x0=np.array([0.0], dtype=np.float64))
print("Minimizer:", res.x)
```

```python
#12) Handles Overflow and Underflow Based on Type.


overflow_arr = np.array([127], dtype=np.int8)
print("Before Overflow:", overflow_arr)
print("After Adding 1:", overflow_arr + 1)  # Wraps to -128
```

## Converting Data Type Using .astype().
The .astype() method lets you convert an array to a new data type safely.

```python
float_arr = np.array([1.1, 2.2, 3.3])
int_arr = float_arr.astype(np.int32)

print("Converted to int:", int_arr)
```

## Memory Efficiency via Smaller Data -  Types.
You can save memory by using smaller types like int8 or float32.

```python
a = np.array([1, 2, 3], dtype=np.int8)
b = np.array([1, 2, 3], dtype=np.int64)

print("Memory used by a:", a.nbytes)
print("Memory used by b:", b.nbytes)
```

## Performing Operations Based on Data -  Type.
Mathematical operations change behavior depending on data type (e.g., integer vs. float).

## Complex Numbers for Scientific Computing.
SciPy supports Complex Numbers using complex64 or complex128.


```python
complex_arr = np.array([1 + 2j, 3 + 4j], dtype=np.complex128)
print("Real Parts:", complex_arr.real)
print("Imaginary Parts:", complex_arr.imag)
```

## Using Boolean Arrays for Masking.
Boolean data types are used for filtering or conditionally operating on data.

```python
arr = np.array([5, 10, 15, 20])
mask = arr > 10

print("Mask:", mask)
print("Filtered Elements:", arr[mask])
```

## Using Structured Data  - Types.
Structured dtypes allow storage of records with named fields (like a database).

```python
dt = np.dtype([('id', 'i4'), ('score', 'f4')])
records = np.array([(1, 95.5), (2, 88.0)], dtype=dt)

print("IDs:", records['id'])
print("Scores:", records['score'])
```

## Finding Size of Elements with .itemsize.
- The .itemsize property shows the size in bytes for one element of the Array.



```python
arr = np.array([10], dtype=np.float64)
print("Bytes per element:", arr.itemsize)
```

## Handling Overflow with Proper Data -  Types.
Wrong data type (like int8) can overflow, use appropriate data  - types to avoid errors.

```python
a = np.array([127], dtype=np.int8)
print("Original:", a)
print("After +1 (Overflow):", a + 1)
```

## SciPy Functions Expect Specific Data - Types.
- Functions like scipy.linalg.inv expect floating-point Inputs, not Integers.



```python
from scipy.linalg import inv

matrix = np.array([[1, 2], [3, 4]], dtype=np.float64)
inverse = inv(matrix)

print("Inverse Matrix:\n", inverse)
```

## Auto Type Promotion in Mixed-Type Operations.
When arrays of different types are used in an operation, NumPy auto-promotes to a higher type.

```python
a = np.array([1], dtype=np.int32)
b = np.array([2.5], dtype=np.float64)
c = a + b

print("Result:", c)
print("Result dtype:", c.dtype)
```
