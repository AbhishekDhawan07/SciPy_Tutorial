## Array Shapes and Dimensions.


- Arrays are the core data structure used in SciPy, built upon NumPy arrays.

- Each array has a shape, which defines the size along each axis (rows, columns, depth, etc.).

- Array dimensions (or ndim) tell how many axes (1D, 2D, 3D…) an array has.

- SciPy uses these shapes/dimensions to apply mathematical operations, filtering, transforms, and more.

- Operations like convolution, linear algebra, Fourier transforms depend on specific array shapes.

- The .shape attribute returns a tuple showing the size of each dimension of an array.

- The .ndim attribute shows the number of dimensions (axes) the array has.

- Reshaping an array changes its structure without changing the data using .reshape() or np.reshape().

- Flattening reduces any n-dimensional array into a 1D array, often used in statistics or ML.

- You can Add/Remove Dimensions using np.expand_dims() and np.squeeze() to match function input requirements.

- Working with shapes and dimensions helps to prepare data correctly for SciPy operations, ensuring accurate results.

- Understanding shapes/dimensions makes SciPy more powerful in image processing, signal processing, data modeling, and scientific computing.

- Array dimensions refer to the number of axes (or levels) an array has, also called its "rank" or .ndim.

- A 1D array is like a simple list (e.g., [1, 2, 3]), a 2D array is like a matrix, and 3D or higher-dimensional arrays are used in images, volumes, and scientific simulations.

- In SciPy, most data is passed around as NumPy arrays, so understanding dimensions is fundamental to using SciPy functions correctly.

- The .ndim property tells you the number of dimensions (axes) the array has.

- You can create arrays of any dimension using np.array() or np.reshape().

- Many SciPy modules like scipy.ndimage, scipy.signal, and scipy.linalg require specific dimensions (e.g., 2D matrices for matrix operations).

- Higher dimensions are useful in batch processing, such as a stack of images (3D), or a sequence of signals (3D/4D arrays).

- Dimension awareness helps you apply correct operations, like summing across a specific axis or filtering a single slice.

- You can add or remove dimensions using np.expand_dims() or np.squeeze() to match expected input formats.

- Flattening (1D conversion) is often used for statistical operations in scipy.stats.

- Functions like np.transpose() or np.swapaxes() help manipulate dimension ordering to match operation needs.

- SciPy’s efficiency comes partly from its ability to work seamlessly with multi-dimensional arrays, performing operations without unnecessary copying.

```python
import numpy as np
from scipy import linalg

# Create a 1D array (vector)
arr1d = np.array([10, 20, 30, 40])
print("1D Array:", arr1d)
print("Shape:", arr1d.shape)       # Output: (4,)
print("Dimensions:", arr1d.ndim)   # Output: 1

# Create a 2D array (matrix).
arr2d = np.array([[1, 2], [3, 4]])
print("\n2D Array:\n", arr2d)
print("Shape:", arr2d.shape)       # Output: (2, 2)
print("Dimensions:", arr2d.ndim)   # Output: 2

# Reshape 1D to 2D (2x2).
reshaped = arr1d.reshape(2, 2)
print("\nReshaped to 2D:\n", reshaped)
print("New Shape:", reshaped.shape)

# Flatten a 2D array to 1D.
flattened = arr2d.flatten()
print("\nFlattened to 1D:", flattened)

# Create a 3D array.
arr3d = np.zeros((3, 2, 4))
print("\n3D Array Shape:", arr3d.shape)
print("3D Array Dimensions:", arr3d.ndim)
```

```python
# Expand a 1D array to 2D (add axis)
arr2d_from_1d = np.expand_dims(arr1d, axis=0)  # Shape: (1, 4)
print("\nExpanded to 2D (1 row):", arr2d_from_1d.shape)

# Squeeze dimensions (remove 1-size dims)
squeezed = np.squeeze(arr2d_from_1d)
print("Squeezed Back to:", squeezed.shape)

# Use shape check before matrix operations
if arr2d.shape[0] == arr2d.shape[1]:
    inv_matrix = linalg.inv(arr2d)
    print("\nInverse of arr2d:\n", inv_matrix)
else:
    print("Matrix is not square!")

# Total number of elements using .size
print("\nTotal Elements in arr2d:", arr2d.size)
```

```python
import numpy as np
from scipy import ndimage, linalg, stats

# 1. Create a 2D array (grayscale image or matrix)
matrix = np.array([[5, 2], [7, 4]])
print("Original Shape:", matrix.shape)  # (2, 2)
print("Dimensions:", matrix.ndim)       # 2D

# 2. SciPy's inverse function requires square matrices
if matrix.shape[0] == matrix.shape[1]:
    inv = linalg.inv(matrix)
    print("\nInverse Matrix:\n", inv)
else:
    print("Not square, cannot invert!")

# 3. Create a 3D image-like array (e.g., RGB image)
image = np.random.rand(100, 100, 3)
print("\nImage Shape:", image.shape)  # (100, 100, 3)

# 4. Convert RGB to grayscale by averaging channels
gray_image = np.mean(image, axis=2)  # Result: (100, 100)
print("Grayscale Shape:", gray_image.shape)

# 5. Apply Gaussian filter to grayscale image
blurred = ndimage.gaussian_filter(gray_image, sigma=1)
print("Blurred Image Shape:", blurred.shape)

# 6. Reshape 1D signal into 2D format for batch processing
signal_1d = np.sin(np.linspace(0, 2*np.pi, 100))
batched_signal = signal_1d.reshape(10, 10)
print("\nBatched Signal Shape:", batched_signal.shape)

# 7. Flatten image for statistical analysis
flat = gray_image.flatten()
print("Flattened for Stats:", flat.shape)

# 8. Compute mean and variance using SciPy
mean = stats.tmean(flat)
var = stats.tvar(flat)
print("Mean:", mean, "| Variance:", var)

# 9. Check dimensions before filtering
if gray_image.ndim == 2:
    sobel_edges = ndimage.sobel(gray_image)
    print("Sobel Edge Detection Done")
else:
    print("Expected 2D input for Sobel filter")

# 10. Use expand_dims to add channel for machine learning
gray_with_channel = np.expand_dims(gray_image, axis=-1)  # (100, 100, 1)
print("Added Channel Dimension:", gray_with_channel.shape)

# 11. Use squeeze to remove singleton dimensions
squeezed_back = np.squeeze(gray_with_channel)
print("Squeezed Back Shape:", squeezed_back.shape)

# 12. Shape guides tiling, broadcasting, and stacking
tile_example = np.tile(signal_1d, (3, 1))  # Shape: (3, 100)
print("\nTiled Signal Shape:", tile_example.shape)
```

## Features of Array Shapes and Dimensions in SciPy.


- Shape Awareness  :   Every SciPy array knows its shape through the .shape attribute, allowing you to programmatically adjust operations to fit the data.

- Dimensional Insight  :  The .ndim property tells how many dimensions (axes) the array has — useful for working with vectors (1D), matrices (2D), or volumes (3D+).

- Automatic Reshaping    :   You can reshape arrays using .reshape() or np.reshape() to prepare them for specific computations (e.g., matrix multiplication, convolution).

- Flexible Flattening    :    Use .flatten() or .ravel() to turn any n-dimensional array into a 1D array, essential for statistical or ML operations in SciPy.

- Broadcasting Support   :   SciPy operations use broadcasting rules that rely on compatible shapes — e.g., adding a 1D array to every row of a 2D array.

- Slicing and Indexing   :   Shape and dimensions allow advanced slicing like extracting rows, columns, or specific depth levels (3D+).

- Dimension Expansion/Reduction   :   Functions like np.expand_dims() and np.squeeze() help prepare data for filters, models, or visualization by adjusting dimensions.

- Compatibility Across Modules   :   Whether using scipy.signal, scipy.ndimage, or scipy.linalg, shape information ensures function compatibility.

- High-dimensional Data Support  :   SciPy can handle arrays with 3, 4, or more dimensions, which is critical in scientific simulations, image stacks, and signal batches.

- Statistical Preprocessing   :   Many SciPy functions (e.g., scipy.stats) require specific shapes (often 1D), so reshaping and flattening are common prep steps.

- Memory Efficiency  :   Shape manipulations (e.g., transpose or reshape) often do not copy the data, making them fast and memory-efficient.

- Visualization-Ready  :   Correct shapes allow easier use of plotting libraries (e.g., matplotlib) to visualize 1D signals, 2D matrices, or 3D volumes.

- Reshaping Arrays  : array.reshape() i.e. Reshapes Array without Changing Data.

- Flattening Arrays   : array.flatten() i.e. Flattens to 1D.

- Transposing Arrays  : array.T	i.e. Transposes the Array.

- Expanding Dimensions  :  Adds dimension at specified axis.

- Squeezing Dimensions  :  Removes single-dimensional entries.

-  Broadcasting Example  : Resizes array (pads or trims).



```python
import numpy as np
from scipy import ndimage, linalg, stats

# 1. Shape awareness with .shape
array2d = np.array([[1, 2, 3], [4, 5, 6]])
print("Shape of array2d:", array2d.shape)  # (2, 3)

# 2. Dimensional insight with .ndim
print("Dimensions:", array2d.ndim)         # 2

# 3. Reshape for compatibility
reshaped = array2d.reshape(3, 2)
print("\nReshaped to 3x2:\n", reshaped)

# 4. Flattening for statistical operations
flattened = reshaped.flatten()
print("Flattened:", flattened)
print("Mean:", stats.tmean(flattened))     # Statistical mean

# 5. Broadcasting with compatible shapes
row = np.array([1, 2, 3])
result = array2d + row  # Broadcast row across 2 rows
print("\nBroadcasted Addition:\n", result)

```

```python
# 6. Slicing using shape
print("First column:", array2d[:, 0])  # All rows, column 0

# 7. Expand and squeeze dimensions
expanded = np.expand_dims(array2d, axis=0)  # Shape: (1, 2, 3)
print("\nExpanded shape:", expanded.shape)

squeezed = np.squeeze(expanded)  # Remove single dimension
print("Squeezed shape:", squeezed.shape)

# 8. High-dimensional array (3D volume)
volume = np.random.rand(5, 5, 5)  # Shape: (depth, height, width)
print("3D Volume Shape:", volume.shape)

# 9. Use shape to loop through image slices (for image stack)
for i in range(volume.shape[0]):
    print(f"Slice {i} shape:", volume[i].shape)

# 10. Transpose shape without copying data
transposed = array2d.T
print("\nTransposed:\n", transposed)

# 11. Check shape before applying matrix inverse
square = np.array([[1, 2], [3, 4]])
if square.shape[0] == square.shape[1]:
    inv = linalg.inv(square)
    print("\nInverse Matrix:\n", inv)

# 12. Plotting requires correct shape (optional for visualization use)
import matplotlib.pyplot as plt
signal = np.sin(np.linspace(0, 2*np.pi, 100))
plt.plot(signal)  # Must be 1D
plt.title("Sine Wave")
plt.show()
```

## Understanding Array Dimensions using .ndim Attribute.


- It tells how many axes (or levels) an array has.
- Example: 1D list, 2D matrix, 3D Image Stack.



```python
import numpy as np

# 1D array
arr1 = np.array([1, 2, 3])
print(arr1.ndim)  # 1

# 2D array
arr2 = np.array([[1, 2], [3, 4]])
print(arr2.ndim)  # 2

# 3D array
arr3 = np.random.rand(2, 2, 3)
print(arr3.ndim)  # 3

# 4D array
arr4 = np.random.rand(1, 3, 4, 2)
print(arr4.ndim)  # 4
```

## Checking Array Shape using shape Function.

- It gives the Number of Elements along each Axis.

- Example: Shape of a 2x3 Matrix is (2, 3).

```python
a = np.random.rand(3, 4)
print(a.shape)  # (3, 4)

b = np.zeros((2, 3, 5))
print(b.shape)  # (2, 3, 5)

c = np.array([[1, 2, 3]])
print(c.shape)  # (1, 3)
```

## Flattening Arrays using flatten() and ravel() Functions.


- Converts multi-dimensional arrays to 1D for statistical analysis.

```python
arr = np.array([[1, 2], [3, 4]])

# flatten() returns a copy
flat = arr.flatten()
print(flat)  # [1 2 3 4]

# ravel() returns a view when possible
ravel = arr.ravel()
print(ravel)  # [1 2 3 4]

# Useful before using scipy.stats
from scipy import stats
print(stats.describe(flat))
```

## Reshaping Arrays using reshape() Function.

- Change the layout of the array without changing data.

```python
arr = np.arange(12)

# Reshape to 2D
reshaped = arr.reshape(3, 4)
print(reshaped)

# Reshape to 3D
arr3d = arr.reshape(2, 3, 2)
print(arr3d)

# Shape mismatch error example
# arr.reshape(5, 3)  # ValueError
```

## Expanding Dimensions using expand_dims() Function.

- Add a new axis; e.g., for batch processing in ML.

```python
arr = np.array([1, 2, 3])

# Add new axis at start
expanded = np.expand_dims(arr, axis=0)
print(expanded.shape)  # (1, 3)

# Add new axis at end
expanded2 = np.expand_dims(arr, axis=-1)
print(expanded2.shape)  # (3, 1)

# Useful in SciPy filters
from scipy.ndimage import gaussian_filter
image = np.random.rand(100, 100)
image_batch = np.expand_dims(image, 0)  # (1, 100, 100)
```

## Removing Dimensions using squeeze() Function.

- Removes axes with size 1; helpful after filtering operations.

```python
arr = np.array([[[1, 2, 3]]])
print(arr.shape)  # (1, 1, 3)

squeezed = np.squeeze(arr)
print(squeezed.shape)  # (3,)

# Only remove if size is 1 along axis
# np.squeeze(arr, axis=0)  # Valid
```

## Transpose and Axis Swapping using .transpose Attribute and  swapaxes Function.

- Reorder axes for operations like matrix multiplication or visualization.

```python
arr = np.array([[1, 2], [3, 4]])

# Transpose 2D
print(arr.T)

# Transpose 3D
arr3d = np.random.rand(2, 3, 4)
print(arr3d.transpose(2, 1, 0).shape)

# Swap axes
swapped = np.swapaxes(arr3d, 0, 2)
print(swapped.shape)
```

```python
a = np.array([[1], [2], [3]])  # Shape: (3,1)
b = np.array([10, 20, 30])     # Shape: (3,)

# Broadcast b to match a
c = a + b
print(c)

# Shape alignment error
# np.add(np.array([1, 2]), np.array([[1, 2], [3, 4]]))  # Error
```
