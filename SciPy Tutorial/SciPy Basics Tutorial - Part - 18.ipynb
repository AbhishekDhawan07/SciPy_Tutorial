## Solving Linear Equations in SciPy.


- In Mathematics, solving a Linear Equation System means Finding a Solution x to the Equation A x = b, where A is a Matrix of Coefficients, and b is a known Result Vector.
  
- In SciPy, this task is handled Efficiently using Optimized Numerical Libraries under the hood(LAPACK/BLAS).

  
- The result x should satisfy the equation within floating-point precision.

  
- SciPy includes both dense solvers (scipy.linalg) and sparse solvers (scipy.sparse.linalg) for different types of matrices.

  
- The key is that you provide A and b, and SciPy finds x that solves the system.

```python
# Example : Solving Linear Equations in SciPy.

from scipy.linalg import solve
import numpy as np

A = np.array([[2, 3], [1, 2]])
b = np.array([8, 5])
x = solve(A, b)
print("Solution:", x)
```

## Features of Solving Linear Equations in SciPy.

## Support for Dense Linear Systems.

- SciPy supports solving standard dense linear systems of the form Ax = b, where A is a Dense Square Matrix.
    
- It uses direct methods like LU decomposition, which are robust and accurate for general systems.

- This is done using the scipy.linalg.solve() function, best suited for small to moderate-sized dense matrices.

- This is the most basic and commonly used feature in linear algebra computations.

```python
from scipy.linalg import solve
import numpy as np

A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])
x = solve(A, b)
print("x =", x)

A2 = np.array([[2, -1], [4, 3]])
b2 = np.array([1, 19])
x2 = solve(A2, b2)
print("x2 =", x2)
```

## LU Decomposition for Repeated Solves.

- For solving multiple systems with the same matrix A, LU factorization is more efficient.
    
- SciPy provides lu_factor() and lu_solve() which allow one-time matrix decomposition.
    
- This avoids recalculating LU every time, making it ideal for control systems and simulations.
    
- You factor once, and reuse it across multiple right-hand sides.

```python
from scipy.linalg import lu_factor, lu_solve

A = np.array([[3, 1], [1, 2]])
lu, piv = lu_factor(A)

b1 = np.array([9, 8])
b2 = np.array([10, 11])
x1 = lu_solve((lu, piv), b1)
x2 = lu_solve((lu, piv), b2)

print("x1 =", x1)
print("x2 =", x2)
```

## Matrix Inversion Is Avoided.

- Instead of explicitly computing A⁻¹, SciPy solves systems via decompositions, which is more stable and faster.
    
- This avoids unnecessary floating-point errors and memory usage.
    
- However, if needed, you can still compute the inverse with inv().
    
- But solving Ax = b using solve() is always preferred over inv(A) @ b.

```python
from scipy.linalg import inv

A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])

x_direct = solve(A, b)
x_via_inv = inv(A) @ b

print("Direct:", x_direct)
print("Via inverse:", x_via_inv)
```

```python
from scipy.linalg import solve
import numpy as np

# Example 5 : LU Decomposition.

from scipy.linalg import lu_factor, lu_solve

A = np.array([[3, 1], [1, 2]])
b1 = np.array([9, 8])
b2 = np.array([10, 12])
lu, piv = lu_factor(A)
x1 = lu_solve((lu, piv), b1)
x2 = lu_solve((lu, piv), b2)
print("x1 =", x1)
print("x2 =", x2)


```

```python
from scipy.linalg import solve
import numpy as np


# Example 6 : Cholesky Decomposition.
from scipy.linalg import cho_factor, cho_solve

A = np.array([[4, 2], [2, 3]])
b = np.array([6, 7])
cf = cho_factor(A)
x = cho_solve(cf, b)
print("x =", x)

```

```python
from scipy.linalg import solve
import numpy as np

# Example 7 : Solve with Complex Numbers.

A = np.array([[1 + 2j, 2], [3, 4 + 1j]])
b = np.array([1 + 1j, 2])
x = solve(A, b)
print("x =", x)
```
