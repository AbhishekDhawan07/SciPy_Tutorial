## Aggreate Functions in SciPy.


- Aggregate Functions compute a single summary value from a set of values—commonly used in statistics, machine learning, signal processing, and scientific computing.

- While aggregate functions are mostly from NumPy, they are widely used in SciPy workflows because SciPy builds upon NumPy.



## Sum i.e. np.sum().

- Computes the Total Sum of all Array Elements. It can be used across the whole array or along a specific axis.

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
print("Sum of all elements:", np.sum(a))
print("Sum along columns:", np.sum(a, axis=0))
```

## Mean i.e. np.mean().

- Calculates the Average Value. Essential in Statistics, Normalization, and Central Tendency Analysis.

```python
print("Mean of matrix:", np.mean(a))
print("Mean along rows:", np.mean(a, axis=1))
```

## Median i.e. np.median().

- Returns the middle value in sorted order. Useful for robust statistics, especially when data has outliers.

```python
print("Median of matrix:", np.median(a))
```

## Standard Deviation i.e. np.std().

- Measures how much values deviate from the mean (spread). Common in statistics and probability.

```python
print("Standard deviation:", np.std(a))
```

## Variance i.e. np.var().

- Calculates the average squared deviation from the mean. It is the square of standard deviation.

```python
print("Variance:", np.var(a))
```

## Minimum i.e. np.min() and Maximum i.e.np.max().


- Finds the smallest or largest element. Used in data normalization and thresholding.

```python
print("Minimum value:", np.min(a))
print("Maximum value:", np.max(a))

```

## Index of Minimum and Maximum i.e. np.argmin() and np.argmax().


- Returns the Index(Position) of the Minimum and Maximum Value in a Flattened Array or Along an Axis.


```python
print("Index of min:", np.argmin(a))
print("Index of max:", np.argmax(a))
```

## Cumulative Sum i.e. np.cumsum().

- Returns the cumulative sum (running total). Useful for integration-like problems and financial analysis.

```python
print("Cumulative sum:", np.cumsum(a))
```

## Cumulative Product i.e.np.cumprod().

- Returns the Cumulative Product(running multiplication). Useful in compound growth modeling.

```python
print("Cumulative product:", np.cumprod(a))
```

## Product of Elements  i.e. np.prod().

- Returns the product of all elements and used in Combinatorics and Multiplicative Models.

```python
print("Product of all elements:", np.prod(a))
```

## Percentiles i.e. np.percentile().

- Returns values below which a given percentage of data falls. Used in outlier detection and quantile analysis.

```python
print("25th percentile:", np.percentile(a, 25))
```

## Any and All i.e. np.any() and np.all().

- Checks if any or all elements are True (non-zero). Used in logical conditions and masks.

```python
b = np.array([1, 0, 3])
print("Any non-zero?", np.any(b))
print("All non-zero?", np.all(b))
```

## Efficient Vectorized Computation.

- Aggregate functions like sum(), mean() etc., are Vectorized—Executed without explicit loops. This ensures High Performance on Large Data Arrays.

```python
import numpy as np

data = np.arange(1, 10001)
print("Sum:", np.sum(data))  # Efficient even on large arrays
```

## Axis-wise Aggregation.

- Most Aggregate Functions support the Axis Argument to reduce along Rows or Columns, essential for Multidimensional Data.

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Sum along rows:", np.sum(matrix, axis=1))
print("Mean along columns:", np.mean(matrix, axis=0))
```

## Multiple Aggregation Types Supported.

- Supports a wide range of statistical aggregations: sum(), mean(), min(), max(), median(), std(), var(), etc.

```python
print("Min:", np.min(matrix))
print("Max:", np.max(matrix))
print("Std Dev:", np.std(matrix))
```

## Custom Percentile and Quantile Computation.


- Allows advanced statistical insights with functions like percentile() and quantile().

```python
print("25th Percentile:", np.percentile(matrix, 25))
```

## Cumulative Operations.

- Functions like np.cumsum() and np.cumprod() compute running totals or products, which are important in simulations and financial models.

```python
arr = np.array([1, 2, 3, 4])
print("Cumulative Sum:", np.cumsum(arr))
print("Cumulative Product:", np.cumprod(arr))
```

## Broadcasting-Compatible Aggregations.

- Aggregate functions are compatible with broadcasted arrays, making them useful in element-wise reductions across matching shapes.

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([10, 20])
result = a + b  # Broadcasting
print("Mean after broadcasting:", np.mean(result))
```

## Usable with SciPy Functions.

- Many scipy modules (like scipy.stats, scipy.optimize) use aggregate functions internally for statistical analysis, cost functions, etc.

```python
from scipy.stats import describe

data = np.array([1, 2, 3, 4, 5])
summary = describe(data)
print("Mean from SciPy:", summary.mean)
print("Variance from SciPy:", summary.variance)
```

## Essential for Data Analysis and Machine Learning.

- Aggregates form the backbone of many ML preprocessing steps: mean normalization, variance thresholding, etc.

```python
X = np.array([[1, 2], [3, 4], [5, 6]])
X_mean_normalized = X - np.mean(X, axis=0)
print("Mean-normalized data:\n", X_mean_normalized)
```

## Foundational in Scientific Computation.

- Aggregate Functions like sum(), mean(), std() are core statistical tools used in most scientific, engineering, and AI/ML computations.

```python
import numpy as np
data = np.array([1.5, 2.3, 3.1])
print("Mean value:", np.mean(data))
```

## Used in Statistical Analysis.

- They are crucial for descriptive statistics, measuring central tendency (mean, median), spread (std, var), and distribution (percentiles).

```python
print("Standard deviation:", np.std(data))
print("25th percentile:", np.percentile(data, 25))
```

## Fundamental in Machine Learning Pipelines.


- Preprocessing steps such as feature scaling, normalization, and outlier detection use aggregate functions.

```python
# Feature normalization.
X = np.array([[10, 20], [30, 40]])
X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
print("Normalized:\n", X_norm)
```

## Supports Axis-Based Computation.


- Allows aggregation across rows (axis=1) or columns (axis=0), essential for matrix and tensor operations.

```python
print("Row-wise mean:", np.mean(X, axis=1))
```

## Speeds Up Computations via Vectorization.

- Aggregate Functions are Vectorized, allowing Fast Computations without Explicit Loops.

```python
big_data = np.random.rand(1000000)
print("Efficient sum:", np.sum(big_data))
```

## Essential for Data Cleaning & NaN Handling.

- np.nanmean() and np.nanstd() allow analysis without removing missing data.

```python
data_with_nan = np.array([1.0, np.nan, 2.0, 3.0])
print("Mean ignoring NaN:", np.nanmean(data_with_nan))
```

## Logical Decisions in Data Filtering.


- Functions like np.any() and np.all() assist in applying filters and decisions.


```python
flags = np.array([True, False, True])
print("Any condition met?", np.any(flags))
```
