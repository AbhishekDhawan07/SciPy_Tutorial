## Linear Algebra operations in SciPy.


- In SciPy, a key subpackage for Linear Algebra is scipy.linalg, which extends the capabilities of numpy.linalg by offering more advanced routines. It allows users to Perform Matrix Operations, Solve Systems of Equations, Compute Matrix Factorizations, and more.



## Matrix Multiplication.

- Matrix Multiplication is a core operation in Linear Algebra where the Product of Two Matrices A and B is defined only if the number of columns in 
A equals the number of rows in B.


```python
import numpy as np
from scipy import linalg

A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])

C = A @ B  # or np.dot(A, B)
print("Matrix Product:\n", C)
```

## Solving Linear Systems (Ax = b). 

- Solves the equation Ax = b, A is a matrix and b is a vector or matrix.

```python
A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])
x = linalg.solve(A, b)
print("Solution x:", x)
```

## Matrix Inversion.


- Finds the Inverse A^ - 1 of a square matrix A, such that A. A^-1 = I(Identity Matrix).
  


```python
A = np.array([[1, 2], [3, 4]])
inv_A = linalg.inv(A)
print("Inverse of A:\n", inv_A)
```

## Determinant of a Matrix.

- The determinant is a scalar that summarizes certain properties of the matrix (e.g., invertibility).

```python
A = np.array([[1, 2], [3, 4]])
det_A = linalg.det(A)
print("Determinant of A:", det_A)
```

## Eigenvalues and Eigenvectors.

- Solves Av=λv, where λ is an Eigenvalue and v is its Eigenvector.

```python
A = np.array([[4, -2], [1, 1]])
eigvals, eigvecs = linalg.eig(A)
print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)
```

## LU Decomposition.


- Decomposes a matrix A into L (lower), U (upper), and permutation matrix P, so  PA=LU.

```python
A = np.array([[3, 1], [1, 2]])
P, L, U = linalg.lu(A)
print("P:\n", P)
print("L:\n", L)
print("U:\n", U)
```

## QR Decomposition.


- Factorizes a matrix A into an orthogonal matrix, Q and upper triangular matrix R such that A = QR.

```python
A = np.array([[1, 1], [1, -1], [1, 2]])
Q, R = linalg.qr(A)
print("Q:\n", Q)
print("R:\n", R)
```

## Singular Value Decomposition(SVD).

- Decomposes A into UΣV^T. Useful in dimensionality reduction, pseudoinverse, etc.

```python
A = np.array([[1, 0], [0, -2], [3, 0]])
U, s, Vh = linalg.svd(A)
print("U:\n", U)
print("Singular values:", s)
print("Vh:\n", Vh)
```

## Norms of Vectors and Matrices.

- Norms measure the size or length of a vector/matrix. Common norms: L2 (Euclidean), Frobenius.

```python
v = np.array([3, 4])
matrix = np.array([[1, 2], [3, 4]])

print("L2 norm of vector:", linalg.norm(v))
print("Frobenius norm of matrix:", linalg.norm(matrix, 'fro'))
```

## Pseudo-Inverse (Moore-Penrose).


- Used when a matrix is not square or not invertible. Computes a "best fit" inverse.

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
pinv_A = linalg.pinv(A)
print("Pseudo-inverse:\n", pinv_A)
```

## Cholesky Decomposition.


- Decomposes a symmetric, positive-definite matrix A into LL^T,  used in efficient numerical solving.

```python
A = np.array([[4, 2], [2, 3]])
L = linalg.cholesky(A, lower=True)
print("Cholesky decomposition (L):\n", L)
```

## Checking Matrix Properties(e.g., Rank, Condition Number).


- Rank: Number of linearly independent rows/columns.

- Condition number: Indicates numerical stability.

```python
A = np.array([[1, 2], [3, 4]])
print("Rank of A:", np.linalg.matrix_rank(A))
print("Condition number of A:", linalg.cond(A))
```

## Solving Systems of Equations.

- Real-world problems in engineering, physics, and economics are often reduced to systems of linear equations. Efficient solutions are critical.



```python
from scipy.linalg import solve
import numpy as np

A = np.array([[3, 2], [1, 2]])
b = np.array([5, 5])
x = solve(A, b)
print("Solution x:", x)
```

## Foundation for Machine Learning & AI.

- ML algorithms like linear regression, PCA, and neural networks are built on matrix operations like dot products, inverses, and decompositions.

```python
X = np.array([[1, 1], [1, 2], [1, 3]])
y = np.array([1, 2, 3])
theta = np.linalg.pinv(X.T @ X) @ X.T @ y
print("Theta (parameters):", theta)
```

## Numerical Stability in Computations.


- Functions like solve() are preferred over directly inverting matrices due to better accuracy and stability in floating-point arithmetic.

```python
from scipy.linalg import solve
A = np.array([[1e-10, 1], [1, 1]])
b = np.array([1, 2])
x = solve(A, b)
print("Numerically stable solution:", x)
```

## Simulation and Modeling.


- Modeling physical systems (e.g., electrical circuits, fluid dynamics) involves linear algebra to handle interrelated variables.

```python
# Ohm's Law: V = IR → I = V / R using matrix form
R = np.array([[2, -1], [-1, 2]])
V = np.array([1, 0])
I = solve(R, V)
print("Current vector I:", I)
```

## Dimensionality Reduction(SVD, PCA).

- Used in data compression and visualization. SciPy’s svd() decomposes data for removing redundancy.

```python
from scipy.linalg import svd
A = np.array([[1, 2], [3, 4], [5, 6]])
U, s, Vh = svd(A)
print("Singular values:", s)
```

## Error Minimization and Regression.

- Linear algebra helps find least-squares solutions—core to regression and optimization techniques.

```python
from scipy.linalg import lstsq
A = np.array([[1, 1], [1, 2], [1, 3]])
b = np.array([1, 2, 2])
x, res, rnk, s = lstsq(A, b)
print("Least Squares Solution:", x)
```

## Matrix Decomposition for Efficient Computation.


- Decompositions (LU, QR, Cholesky) simplify solving systems, inverting, and finding eigenvalues more efficiently.

```python
from scipy.linalg import lu
A = np.array([[3, 1], [1, 2]])
P, L, U = lu(A)
print("L:\n", L)
print("U:\n", U)
```
